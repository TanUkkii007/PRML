\mathbf{\Lambda}_{aa} & \mathbf{\Lambda}_{ab} & \mathbf{\Lambda}_{ac}\\
\mathbf{\Lambda}_{ba} & \mathbf{\Lambda}_{bb} & \mathbf{\Lambda}_{bc}\\
\mathbf{\Lambda}_{ca} & \mathbf{\Lambda}_{cb} & \mathbf{\Lambda}_{cc}
\end{pmatrix}$

---

同時分布$p(\mathbf{x}_a, \mathbf{x}_b, \mathbf{x}_c)$の周辺分布$p(\mathbf{x}_a, \mathbf{x}_b)$の平均と共分散は、(2.98)より同時分布の平均と分散から周辺化する変数を消去した部分行列となる。

$\mathbf{\mu}_{a, b} = \begin{pmatrix}
\mathbf{\mu}_a\\
\mathbf{\mu}_b
\end{pmatrix}$

$\mathbf{\Sigma}_{a, b} = \begin{pmatrix}
\mathbf{\Sigma}_{aa} & \mathbf{\Sigma}_{ab}\\
\mathbf{\Sigma}_{ba} & \mathbf{\Sigma}_{bb}
\end{pmatrix}$

---

$(\mathbf{x} - \mathbf{\mu})^\top \Sigma^{-1}(\mathbf{x} - \mathbf{\mu}) = \\
(\mathbf{x}_a - \mathbf{\mu}_a)^\top \Lambda_{aa}(\mathbf{x}_a - \mathbf{\mu}_a) + (\mathbf{x}_a - \mathbf{\mu}_a)^\top \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) + (\mathbf{x}_a - \mathbf{\mu}_a)^\top \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)\\
+ (\mathbf{x}_b - \mathbf{\mu}_b)^\top \Lambda_{ba}(\mathbf{x}_a - \mathbf{\mu}_a) + (\mathbf{x}_b - \mathbf{\mu}_b)^\top \Lambda_{bb}(\mathbf{x}_b - \mathbf{\mu}_b) +  (\mathbf{x}_b - \mathbf{\mu}_b)^\top \Lambda_{bc}(\mathbf{x}_c - \mathbf{\mu}_c)\\
+ (\mathbf{x}_c - \mathbf{\mu}_c)^\top \Lambda_{ca}(\mathbf{x}_a - \mathbf{\mu}_a) + (\mathbf{x}_c - \mathbf{\mu}_c)^\top \Lambda_{cb}(\mathbf{x}_b - \mathbf{\mu}_b) + (\mathbf{x}_c - \mathbf{\mu}_c)^\top \Lambda_{cc}(\mathbf{x}_c - \mathbf{\mu}_c)$

$p(\mathbf{x}_a|\mathbf{x}_b, \mathbf{x}_c)$を考える。$\mathbf{x}_b, \mathbf{x}_c$を定数とみなす。

$\mathbf{x}_a$の２次項は$(\mathbf{x}_a - \mathbf{\mu}_a)^\top \Lambda_{aa}(\mathbf{x}_a - \mathbf{\mu}_a)$なので、共分散$\Sigma_{a|b, c}$は、

$\Sigma_{a|b, c} = \Lambda_{aa}^{-1}$

$\mathbf{x}_a$の1次項は、

$-\mathbf{x}_a^\top \Lambda_{aa}\mathbf{\mu}_a - \mathbf{\mu}_a^\top \Lambda_{aa}\mathbf{x}_a + \mathbf{x}_a^\top \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) + \mathbf{x}_a^\top \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)\\
+ (\mathbf{x}_b - \mathbf{\mu}_b)^\top \Lambda_{ba}\mathbf{x}_a\\
+ (\mathbf{x}_c - \mathbf{\mu}_c)^\top \Lambda_{ca}\mathbf{x}_a = $

$-\mathbf{x}_a^\top \Lambda_{aa}\mathbf{\mu}_a - \mathbf{x}_a^\top\Lambda_{aa}^\top\mathbf{\mu}_a + \mathbf{x}_a^\top \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) + \mathbf{x}_a^\top \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)\\
+ \mathbf{x}_a^\top\Lambda_{ba}^\top(\mathbf{x}_b - \mathbf{\mu}_b)\\
+ \mathbf{x}_a^\top\Lambda_{ca}^\top (\mathbf{x}_c - \mathbf{\mu}_c)$ 　　　２次形式は対称なので、転置をとっても等価であるから

$ = -\mathbf{x}_a^\top \Lambda_{aa}\mathbf{\mu}_a - \mathbf{x}_a^\top\Lambda_{aa}\mathbf{\mu}_a + \mathbf{x}_a^\top \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) + \mathbf{x}_a^\top \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)\\
+ \mathbf{x}_a^\top\Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b)\\
+ \mathbf{x}_a^\top\Lambda_{ac} (\mathbf{x}_c - \mathbf{\mu}_c)$ 　　　$\Lambda_{aa} = \Lambda_{aa}^\top, \Lambda_{ab} = \Lambda_{ba}^\top, \Lambda_{ac} = \Lambda_{ca}^\top$より

$ = -2\mathbf{x}_a^\top \Lambda_{aa}\mathbf{\mu}_a + 2\mathbf{x}_a^\top \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) + 2\mathbf{x}_a^\top \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)$

$ = -2\mathbf{x}_a^\top \left(\Lambda_{aa}\mathbf{\mu}_a - \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) - \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)\right)$

よって、

$\mathbf{\Sigma}_{a|b,c}^{-1}\mathbf{\mu}_{a|b, c} = \Lambda_{aa}\mathbf{\mu}_a - \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) - \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)$

よって平均$\mathbf{\mu}_{a|b, c}$は、

$\mathbf{\mu}_{a|b, c} = \mathbf{\Sigma}_{a|b,c}\left(\Lambda_{aa}\mathbf{\mu}_a - \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) - \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)\right)$

$ = \Lambda_{aa}^{-1}\left(\Lambda_{aa}\mathbf{\mu}_a - \Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) - \Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)\right)$ 　　　$\Sigma_{a|b, c} = \Lambda_{aa}^{-1}$より、

$ = \mathbf{\mu}_a - \Lambda_{aa}^{-1}\Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b) - \Lambda_{aa}^{-1}\Lambda_{ac}(\mathbf{x}_c - \mathbf{\mu}_c)$

(2.97), (2.98)より、周辺分布の平均は条件付き分布の平均から消去する条件の変数の項を取り除くことで得られる。よって、

$\mathbf{\mu}_{a|b} = \mathbf{\mu}_a - \Lambda_{aa}^{-1}\Lambda_{ab}(\mathbf{x}_b - \mathbf{\mu}_b)$

また共分散は(2.95), (2.96)より、同時分布の精度行列の該当する変数の部分行列の逆行列となるので、

$\mathbf{\Sigma}_{a|b} = \mathbf{\Lambda}_{aa}^{-1}$
